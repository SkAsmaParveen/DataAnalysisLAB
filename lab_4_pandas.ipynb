```python
from datetime import datetime , date

# Current_Age = Date of Birth - Today's Date

# To identify the date present in the variable we used strptime function

age_of_purchase = datetime.strptime(df.purhcase_date[0] , "%d/%m/%Y").date()
print(age_of_purchase)

#Get today's date 
today = date.today() 
print(today)

print("Age_of_Purchase :", 
      today.year - age_of_purchase.year - ((today.month, 
                                          today.day) < (age_of_purchase.month, 
                                                        age_of_purchase.day)))
```
